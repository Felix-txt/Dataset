```python
import pandas as pd

df = pd.read_csv('Google_Stock_Price.csv')

df
```

```python
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'])

df.set_index('date', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df['close'], label='Google Stock Price (Close)', color='blue')
plt.title('Google Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['open'], label='Google Stock Price (Open)', color='red')
plt.title('Google Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['high'], label='Google Stock Price (High)', color='red')
plt.plot(df['low'], label='Google Stock Price (Low)', color='blue')
plt.title('Google Stock Price High and Low')
plt.xlabel('date')
plt.ylabel('Stock Price (USD)')
plt.grid()
plt.show()
```

```python
ds = df.drop(['dividends', 'splits', 'symbol'], axis=1)
ds
```

```python
import plotly.graph_objs as go

ds.columns = ds.columns.str.strip()

ds['date'] = pd.to_datetime(ds['date'])

ds = ds[(ds['date'] >= '2018-01-01') & (ds['date'] <= '2025-12-31')]

# Skapa grafen
fig = go.Figure()

# LÃ¤gg till linje fÃ¶r justerade stÃ¤ngningspriser
fig.add_trace(go.Scatter(x=ds['date'], y=ds['adjclose'], mode='lines', name='Adj Close'))

# Anpassa layout
fig.update_layout(title='Google Stock Price (2018-2025)',
                  xaxis_title='Datum',
                  yaxis_title='Price (USD)',
                  hovermode='closest')

# Visa grafen
fig.show()
```
